```python
from qiskit import *
import matplotlib.pyplot as plt
```

```python
a = QuantumCircuit(2)
a.h(0)
a.cx(0, 1)
a.barrier()
a.x(0)
a.z(0)
a.barrier()
a.cx(0, 1)
a.h(0)
a.draw(output='mpl', initial_state='True')
```

```python
from qiskit.quantum_info import Statevector
sva = Statevector.from_instruction(a)
sva.draw('latex')
```
