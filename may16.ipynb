```python
import qiskit, pylatexenc
```

```python
from qiskit import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
oracle = QuantumCircuit(2, name='Oracle for 11')
oracle.cz(0, 1)
oracle.to_gate()
oracle.draw(output='mpl')
```

```python
grover_circ = QuantumCircuit(2)
grover_circ.h([0, 1])
grover_circ.append(oracle, [0, 1])
grover_circ.draw(output='mpl')
```

```python
reflection = QuantumCircuit(2, name='Reflection Circuit')
reflection.h([0, 1])
reflection.z([0, 1])
reflection.cz(0, 1)
reflection.h([0, 1])
reflection.to_gate()
```

```python
reflection.draw('mpl')
```

```python
grover_circ.append(reflection, [0, 1])
grover_circ.draw('mpl', initial_state=True)
```

```python
from qiskit.quantum_info import Statevector
sv = Statevector.from_instruction(grover_circ)
sv.draw('latex')
```
