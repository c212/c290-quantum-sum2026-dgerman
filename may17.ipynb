```python
import qiskit
```

```python
from qiskit import *
```

```python
circuit = QuantumCircuit(6)
circuit.x(5)
circuit.barrier()
circuit.h(range(6))
circuit.barrier()
circuit.cx(2, 5)
circuit.cx(4, 5)
circuit.cx(0, 5)
circuit.barrier()
circuit.h(range(6))
circuit.draw(output='mpl')
```

```python
from qiskit.quantum_info import Statevector
sv = Statevector.from_instruction(circuit)
sv.draw('latex')
```
