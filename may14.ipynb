```python
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(2)
qc.x(0)
qc.x(1)
qc.barrier()
qc.h(0)
qc.h(1)
qc.barrier()
qc.x(0)
qc.cx(0, 1)
qc.x(0)
qc.barrier()
qc.h(0)
qc.h(1)
display(qc.draw('mpl', initial_state=True))
```

```python
from qiskit.quantum_info import Statevector
sv = Statevector.from_instruction(qc)
sv.draw('latex')
```
